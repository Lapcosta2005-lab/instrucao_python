# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)

print(vendas)
print(meses)
```

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.show
```

- Editando o Gráfico de Linha

```python
# Mudando para marcadores ao invés de linhas
plt.plot(meses, vendas, 'ro') # Adicionado o 'ro'
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.show
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
plt.plot(meses, vendas, 'ro', color='blue') # Adicionado cor
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.show
```

- Dispersão

```python
plt.scatter(meses,vendas)
plt.show
```

- Barras

```python
plt.bar(meses,vendas, color="green")
plt.show
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1) # Figura número 1

plt.subplot(1, 3, 1) 
# Figura de 1 linha, 1 coluna com o primeiro gráfico na primeira posição
plt.plot(meses, vendas)

plt.subplot(1, 3, 2)
plt.bar(meses,vendas, color="green")

plt.subplot(1, 3, 3)
plt.scatter(meses,vendas)

plt.show()
```

```python
plt.figure(1, figsize=(20,5)) # Acrescentado o tamanho da figura (20 de alargura e 5 de altura)

plt.subplot(1, 3, 1) 
# Figura de 1 linha, 1 coluna com o primeiro gráfico na primeira posição
plt.plot(meses, vendas)

plt.subplot(1, 3, 2)
plt.bar(meses,vendas, color="green")

plt.subplot(1, 3, 3)
plt.scatter(meses,vendas)

plt.show()
```
