# Como instalar um Módulo/Biblioteca que não tenha no seu computador

### Estrutura:

- Usaremos o "pip" para isso. É um gerenciador de módulos/pacotes/bibliotecas que ajuda a gente a instalar, atualizar, desinstalar novos módulos
- Não precisa dominar pip, apenas os comandos que vamos usar aqui mesmo

Passo 1: Abrir o Prompt do Anaconda<br>
Passo 2: Instalar com o pip os pacotes que quiser<br>
Passo 3: Reiniciar o Jupyter (nem sempre necessário, mas na dúvida é melhor fazer)<br>

```python
import keyboard
import time
import pyautogui
import pygetwindow as gw # Importa pygetwindow com um alias para facilitar

# --- Configurações ---
# Caminho para a imagem do perfil "Pessoa 1" que você capturou
# Lembre-se de usar 'r' para raw string para caminhos do Windows
IMAGE_PATH_PESSOA1 = r'C:\Users\SEINT-RS\Downloads\pessoa1.png'

# Tempos de espera ajustáveis
TEMPO_ESPERA_CURTO = 1   # Para reações rápidas do sistema (ex: abrir menu Win)
TEMPO_ESPERA_MEDIO = 2   # Para ações como abrir nova janela ou digitar texto
TEMPO_ESPERA_LONGO = 5   # Para carregar aplicativos ou páginas web complexas

URL_DESTINO = "http://www.terra.com.br" # Definindo a URL como uma variável para fácil modificação

# --- Automação ---
try:
    # 1. Abre a barra de pesquisa (Menu Iniciar)
    print("Pressionando tecla Win...")
    keyboard.press_and_release('win')
    time.sleep(TEMPO_ESPERA_CURTO) # Espera o menu/barra de pesquisa aparecer

    # 2. Digita 'chrome'
    print("Digitando 'chrome'...")
    keyboard.write('chrome')
    time.sleep(TEMPO_ESPERA_CURTO) # Espera o sistema encontrar o Chrome

    # 3. Pressiona Enter para abrir o Chrome
    print("Pressionando Enter para abrir o Chrome...")
    keyboard.press_and_release('enter')
    time.sleep(TEMPO_ESPERA_LONGO) # Espera o Chrome abrir e a tela de perfis aparecer

    # 4. Tenta encontrar e focar na janela do Chrome (para maior robustez)
    chrome_window_found = False
    try:
        chrome_windows = gw.getWindowsWithTitle('Google Chrome') # Pode precisar ajustar o título exato da janela
        if chrome_windows:
            chrome_window = chrome_windows[0] # Pega a primeira janela encontrada
            if not chrome_window.isActive:
                 print("Focando na janela do Chrome...")
                 chrome_window.activate() # Ativa a janela
                 time.sleep(TEMPO_ESPERA_CURTO) # Espera um pouco depois de ativar
            chrome_window_found = True
        else:
            print("Aviso: Janela 'Google Chrome' não encontrada. A automação pode falhar.")
    except Exception as e:
        print(f"Erro ao tentar focar na janela do Chrome: {e}. Tentando continuar sem focar.")

    # Se a janela não foi encontrada, pode não ser útil continuar com o clique de imagem
    if not chrome_window_found:
        print("Não foi possível garantir que a janela do Chrome está ativa. Parando o script.")
        # Opcional: raise Exception("Chrome window not found or activated.") para parar aqui.
        # Opcional: sys.exit() # Para sair do programa completamente
    else:
        # 5. Tenta localizar a imagem do perfil "Pessoa 1" na tela e clicar nela
        print(f"Procurando pela imagem do perfil '{IMAGE_PATH_PESSOA1}' na tela...")
        # Certifique-se de que opencv-python está instalado para usar 'confidence'
        local_pessoa1 = pyautogui.locateCenterOnScreen(IMAGE_PATH_PESSOA1, confidence=0.9)

        if local_pessoa1 is not None:
            # Se a imagem foi encontrada, clique nas coordenadas do centro
            print(f"Perfil 'Pessoa 1' encontrado em {local_pessoa1}. Clicando...")
            pyautogui.click(local_pessoa1)
            print("Perfil selecionado.")
            time.sleep(TEMPO_ESPERA_LONGO) # Espera o perfil carregar completamente (página inicial)

            # --- NOVOS COMANDOS AQUI ---

            # 6. Abre uma nova janela do Chrome (atalho Ctrl+N)
            print("Abrindo nova janela do Chrome (Ctrl+N)...")
            keyboard.press_and_release('ctrl+n')
            time.sleep(TEMPO_ESPERA_MEDIO) # Espera a nova janela abrir e focar na barra de endereço

            # 7. Digita o link na barra de endereços
            print(f"Digitando o link '{URL_DESTINO}'...")
            keyboard.write(URL_DESTINO)
            time.sleep(TEMPO_ESPERA_CURTO) # Pequena pausa após digitar o link

            # 8. Pressiona Enter para navegar para o link
            print("Pressionando Enter para navegar...")
            keyboard.press_and_release('enter')
            time.sleep(TEMPO_ESPERA_LONGO) # Espera a página do Terra carregar

            print(f"Link '{URL_DESTINO}' aberto com sucesso na nova janela.")

            # 9. Colocando a janela em tela inteira (atalho Function+F11)
            print("Colocando a janela em tela inteira (atalho Function+F11)...")
            keyboard.press_and_release('F11')
            time.sleep(TEMPO_ESPERA_MEDIO) # Espera a nova janela abrir e focar na barra de endereço




        else:
            print(f"Erro: Imagem do perfil '{IMAGE_PATH_PESSOA1}' não encontrada na tela.")
            print("Verifique se o Chrome está aberto na tela correta, a resolução da tela e se a imagem é fiel.")

except FileNotFoundError:
    print(f"Erro: O arquivo de imagem '{IMAGE_PATH_PESSOA1}' não foi encontrado.")
    print("Certifique-se de que o caminho da imagem está correto e que o arquivo existe.")

except pyautogui.PyAutoGUIException as e:
     print(f"Erro relacionado a PyAutoGUI (certifique-se de que Pillow e opencv-python estão instalados): {e}")

except Exception as e:
    print(f"Ocorreu um erro inesperado durante a automação: {e}")

print("\nScript de automação concluído (ou encontrou um erro).")
```
