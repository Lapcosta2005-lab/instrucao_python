```python
# 1º) Baixar a biblioteca "requests"
#pip install requests

import requests

# 2º) buscar a API_KEY
api_key = "c5ded0f21ca54fab84a122730251804"

# 3º) pegar o link URL da API através de uma REQUEST (Requisição)
# URL_base = "http://api.weatherapi.com/v1" + clima_atual = "/current.json"
# URL = "http://api.weatherapi.com/v1/current.json"
link_api = "http://api.weatherapi.com/v1/current.json"
link_api_previsao = "http://api.weatherapi.com/v1/forecast.json"

# 4º) Para retirar o erro 401, devem ser repassados alguns parâmetros (request parameters), como 
# - chave da API (key)
# - Parâmetro do local a ser consultado (q)
cidade = input("Qual cidade você quer saber o clima atual? ")
dias = input("Para quantos dias você quer a previsão? ")
parametros = {
    "key": api_key, 
    "q": cidade, 
    "lang": "pt",
    "days": dias}

# Agora vamos coletar (get) as informações do link e armazenar na variável "resposta"
resposta = requests.get(link_api, params=parametros)
resposta2 = requests.get(link_api_previsao, params=parametros)

print(resposta.status_code)
print(resposta.content) # Resposta: API inválida ou não fornecida

print(resposta2.status_code)
print(resposta2.content)
```

# STATUS CODE:
## resposta 401 --> Não satisfez a requisição (Neste caso, sem autorização)
No nosso exemplo, temos que passar pleo menos as requisições obrigatórias --> "KEY" e "Q" e opcionais (LANG)
## resposta 200 --> Deu certo a requisição
# resposta 300 --> Requisição direcionada
# resposta 500 --> servidor fora do ar


```python
if resposta2.status_code == 200:
    dados_requisicao = resposta2.json() # Transforma a resposta json em um dicionario python
    print(dados_requisicao)
```
